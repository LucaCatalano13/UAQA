```python
import rasterio
import json

raster = rasterio.open("milan_sample_dataset/dem/2023-09-30.tiff")
layers = json.load( open("milan_sample_dataset/dem/layers.json"))
type(raster)
```

```python
raster.meta
```

```python
array = raster.read()
print(f"raster.read() generates a {type(array)} of shape:\n(ch = {array.shape[0]}, height = {array.shape[1]}, width = {array.shape[2]})")
```

```python
from rasterio.plot import show
from rasterio.plot import show_hist

for i , band in enumerate(array):
    print(f"channel {i}")
    print(f"name:\t{layers[i]}")
    print(f"min:\t{band.min()}")
    print(f"max:\t{band.max()}")
    show((raster,i+1))
    #sns.heatmap(band)

show_hist(raster)
```
